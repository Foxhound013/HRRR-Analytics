```python
# This script is intended to cat together all X hour forecasts from each month so that there will be a continuous Xhr forecast file for December through March.
# Files need to all be in a common site folder.
import os
import pandas as pd
```

```python
#common_path = 'C:/Users/wdownin/Desktop/backup/hrrr_closest2rwis_processed/all_out'
common_path = 'C:/Users/wdownin/Desktop/backup2/hrrr_closest2rwis_processed/all_out'
```

```python
files = os.listdir(common_path)
files
```

```python
sites = os.listdir(common_path)
sites
```

```python
hours = ['00', '01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12', '13', '14', '15', '16', '17', '18']
for hr in hours:
    fileClue = 'fcsthr' + hr
    out = '/master/dec-march'+hr+'.csv'

    for site in sites:
        print("Operating on: " + site)
        files = os.listdir(common_path+'/'+site+'/')
        fcstHrs = [file for file in files if fileClue in file] # Grab the filename from the current directory

        # Make certain that the file doesn't exist yet, if it does, remove it.
        try:
            os.remove(common_path+'/'+site+out)
            print('The file already existed, it has been removed so as to be updated.')
        except:
            print("The file hasn't been created yet. Beginning creation now. . .")

        header_flag=True
        for fcstHr in fcstHrs:
            print('Processing forecast hour: ' + fcstHr)
            csv = open(common_path+'/'+site+out, 'a')

            f = open(common_path+'/'+site+'/'+fcstHr)
            if header_flag:
                for line in f:
                    csv.write(line)
                f.close()
                header_flag = False
            else:
                f.__next__() # skip the header
                for line in f:
                    csv.write(line)
                f.close()

        csv.close()

    
```
